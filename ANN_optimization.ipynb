```python
import pandas as pd
import os
import numpy as np
import warnings
warnings.filterwarnings("ignore")
def load_conduct_data(conduct_path="E:/code"):
    csv_path = os.path.join(conduct_path, "foam1.csv")
    return pd.read_csv(csv_path)
import keras.backend as K
def r2(y_test, y_pred):
    a = K.square(y_pred - y_test)
    b = K.sum(a)
    c = K.mean(y_test)
    d = K.square(y_test - c)
    e = K.sum(d)
    f = 1 - b/e
    return f
```

```python
#将在特征选择中得到的特征填入final_list中
final_list = ['VF2', 'GXD', 'SVF', 'GMAXVX', 'GMINVX', 'GMAXVY', 'MEANCV', 'MPDV3', 'MPDM3', 'MINCV', 'GMAXVZ', 'MPDM2', 'D1', 'MPDV1', 'MCLZY2', 'MCLXZ2', 'MCLZX2', 'MCLXYZ2', 'MCLXZY2', 'MCLYZX2', 'GSDY', 'GSDX', 'MCLZ2', 'MCLY2', 'MCLX2', 'MCLXY2', 'MCLXZ1', 'MCLYZ1', 'MCLYZX1', 'MCLXZY1', 'MCLYX1', 'V1', 'SA1', 'SSA1', 'SA2', 'OMAXTL', 'CP', 'SIM2', 'LBS', 'OMEANTL', 'OMINTL', 'NOVEXD', 'NOVEYD', 'NOCC', 'SAE', 'SSAE', 'NOVS', 'SSAV', 'IT1', 'SIT1', 'EC62', 'L1', 'L2', 'PLV3', 'LBOD1', 'OD1', 'HBOD1', 'ST', 'SIM1', 'UBS', 'SorD', 'POST', 'MAXCV', 'Mass', 'GMVX', 'GMINVY', 'D3', 'GSDZ', 'IM2']
print(len(final_list))
c_c = np.array(conduct[final_list])
c_X = c_c
c_y = c[:,131:132]
c_X = preprocessing.scale(c_X)
c_X_train_full, c_X_test, c_y_train_full, c_y_test = train_test_split(
   c_X, c_y,random_state = 32, shuffle = True)
c_X_train, c_X_valid, c_y_train, c_y_valid = train_test_split(
    c_X_train_full, c_y_train_full,random_state = 32, shuffle = True)
input_ = keras.layers.Input(shape=c_X_train.shape[1:])
```

```python
# optimization for ANN with 2 hidden layers
mae_final = 100
index = 0
for d1 in range(10,50):
    for d2 in range(10,50):  
        index = index + 1
        #d1,d2表示神经元的数量，activation是使用的激活函数，可以根据需求替换为别的激活函数
        hidden1 = keras.layers.Dense(d1, activation=LeakyReLU(alpha=0.3))(input_)
        hidden2 = keras.layers.Dense(d2, activation=LeakyReLU(alpha=0.3))(hidden1)
        concat = keras.layers.concatenate([input_, hidden2])
        output = keras.layers.Dense(1)(concat)
        model = keras.models.Model(inputs=[input_], outputs=[output])
        #提前停止训练，如果在1000次的训练中模型的误差不再降低通常说明模型已经收敛，patience可以根据实际情况改变
        callback = keras.callbacks.EarlyStopping(monitor='loss', patience=1000, mode = 'min')
        #optimizer为优化器，可根据需求进行改变，loss是我们最关注的误差指标，可以根据需求替换为r2、mse、mae等
        model.compile(loss='mape',optimizer=keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08),metrics=['mse',r2])
        #c_X_train、c_y_train是训练集的特征和标签，epochs是训练的次数
        history = model.fit(c_X_train, c_y_train, epochs=30000,verbose = 0,callbacks=[callback])
        mse_test = model.predict(c_X_test)
        #得到本轮训练的误差
        mae_cur = mean_absolute_percentage_error(c_y_test, mse_test)
        if(index % 10 == 0):
            print(f"cuurent_index={index}")
        #如果当前误差是最小的，则打印出来
        if(mae_cur < mae_final):
            mae_final = mae_cur
            f1 = d1
            f2 = d2
            print(f"note note note mae={mae_final},f1={f1},f2={f2}")
```

```python
# optimization for ANN with 3 hidden layers
from keras import regularizers
mae_final = 100
index = 0
for d1 in range(10,30):
    for d2 in range(10,30):  
        for d3 in range(10,30): 
            index = index + 1
            #d1,d2,d3表示神经元的数量，activation是使用的激活函数，可以根据需求替换为别的激活函数
            hidden1 = keras.layers.Dense(d1, kernel_regularizer=regularizers.l2(0.01), activation=LeakyReLU(alpha=0.3))(input_)
            hidden2 = keras.layers.Dense(d2, kernel_regularizer=regularizers.l2(0.01),activation=LeakyReLU(alpha=0.3))(hidden1)
            hidden3 = keras.layers.Dense(d3, kernel_regularizer=regularizers.l2(0.01),activation=LeakyReLU(alpha=0.3))(hidden2)
            concat = keras.layers.concatenate([input_, hidden2,hidden3])
            output = keras.layers.Dense(1)(concat)
            model = keras.models.Model(inputs=[input_], outputs=[output])
            #提前停止训练，如果在1000次的训练中模型的误差不再降低通常说明模型已经收敛，patience可以根据实际情况改变
            callback = keras.callbacks.EarlyStopping(monitor='loss', patience=1000, mode = 'min')
            #optimizer为优化器，可根据需求进行改变，loss是我们最关注的误差指标，可以根据需求替换为r2、mse、mae等
            model.compile(loss='mape',optimizer=keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08),metrics=['mse',r2])
            #c_X_train、c_y_train是训练集的特征和标签，epochs是训练的次数
            history = model.fit(c_X_train, c_y_train, epochs=30000,verbose = 0,callbacks=[callback])
            mse_test = model.predict(c_X_test)
            #得到本轮训练的误差
            mae_cur = mean_absolute_percentage_error(c_y_test, mse_test)
            if(index % 10 == 0):
                print(f"cuurent_index={index}")
            #如果当前误差是最小的，则打印出来
            if(mae_cur < mae_final):
                mae_final = mae_cur
                f1 = d1
                f2 = d2
                f3 = d3
                print(f"note note note mae={mae_final},f1={f1},f2={f2},f3={f3}")
```
